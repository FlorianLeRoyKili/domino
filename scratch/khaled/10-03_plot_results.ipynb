```python
from domino.plot import sdm_barplot, sdm_displot, generate_group_df
import matplotlib.pyplot as plt
import numpy as np
```

## Plot embedding comparison

```python
synthetic = False
if synthetic:
    score_sdm_ids=[6634]
else:
    score_sdm_ids=[5353]

sdm_barplot(
    score_sdm_ids=score_sdm_ids,
    emb_groups=["eeg", "multimodal"],
    sdm_classes=["domino.sdm.gmm.MixtureModelSDM"],
    run_dir="."
)

#plt.savefig(f"figures/emb_barplot_synthetic={synthetic}.pdf")
plt.show()

sdm_displot(
    score_sdm_ids=score_sdm_ids, #synthetics: 3918
    emb_groups=["eeg", "multimodal"],
    sdm_classes=["domino.sdm.gmm.MixtureModelSDM"],
)
#plt.savefig(f"figures/emb_displot_synthetic={synthetic}.pdf")
plt.show()
```

## Plot sdm comparison

```python
out = sdm_barplot(
    score_sdm_ids=score_sdm_ids,
    emb_groups=["multimodal"],

    sdm_classes=[
        "domino.sdm.gmm.MixtureModelSDM",
        "domino.sdm.george.GeorgeSDM",
        "domino.sdm.multiaccuracy.MultiaccuracySDM",
        "domino.sdm.spotlight.SpotlightSDM",
        "domino.sdm.confusion.ConfusionSDM",
    ],
    hue="sdm_class"
)

plt.savefig(f"figures/sdm_barplot_synthetic={synthetic}.pdf")
plt.show()

sdm_displot(
    score_sdm_ids=score_sdm_ids, #synthetics: 3918
    emb_groups=["multimodal"],
        sdm_classes=[
        "domino.sdm.gmm.MixtureModelSDM",
        "domino.sdm.george.GeorgeSDM",
        "domino.sdm.multiaccuracy.MultiaccuracySDM",
        "domino.sdm.spotlight.SpotlightSDM",
        "domino.sdm.confusion.ConfusionSDM",
    ],
    hue="sdm_class"
)
plt.savefig(f"figures/sdm_displot_synthetic={synthetic}.pdf")
plt.show()
```
